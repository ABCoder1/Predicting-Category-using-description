```python
case1_dataset = case1_dataset.dropna()
```

```python
import nltk
case1_dataset['Description_Tokens'] = case1_dataset.apply(lambda row: nltk.word_tokenize(str(row['description'])), axis=1)
```

```python
case1_dataset[]
```

```python
# Removing extra characters from the product_category_tree
case1_dataset["product_category_tree"] = case1_dataset["product_category_tree"].map(lambda x: x.lstrip('["').rstrip('"]'))
```

```python
case1_dataset
```

```python
from anytree import Node, RenderTree, PreOrderIter, find_by_attr

# Creating the Category Tree as Python Data structure - Anytree

for ind in case1_dataset.index :
    lines = case1_dataset.loc[ ind, "Product_category_tree"]

    root = Node(lines[0].split(",")[1].strip())

    for line in lines:
        
        line = line.split(",")[1]
            
        for idx, cats in enumerate(line.split("^")[1:]):
            
        catNode = find_by_attr(root, cats.strip())
            
        if (catNode is None):
            parentNode = find_by_attr(root, line.split("^")[idx])
    
            if parentNode is None:
                Node(cats.strip(), parent=root)
            else:
                Node(cats.strip(), parent=parentNode)

            
for pre, _, node in RenderTree(root):
                print("%s%s" % (pre, node.name))
                
```
